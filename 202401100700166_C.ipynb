Case Study 4 : To do list Builder

```python
tasks = []

def add_task():
  task = input("Enter the task: ")
  tasks.append({"task": task, "completed": False})
  print("Task added successfully!")

def delete_task():
  view_list()
  try:
    task_index = int(input("Enter the index of the task to delete: ")) - 1
    if 0 <= task_index < len(tasks):
      del tasks[task_index]
      print("Task deleted successfully!")
    else:
      print("Invalid task index.")
  except ValueError:
    print("Invalid input. Please enter a number.")

def view_list():
  if not tasks:
    print("Your to-do list is empty.")
    return

  print("Your to-do list:")
  for index, task in enumerate(tasks):
    status = "[x]" if task["completed"] else "[ ]"
    print(f"{index + 1}. {status} {task['task']}")

def mark_complete():
  view_list()
  try:
    task_index = int(input("Enter the index of the task to mark as complete: ")) - 1
    if 0 <= task_index < len(tasks):
      tasks[task_index]["completed"] = True
      print("Task marked as complete!")
    else:
      print("Invalid task index.")
  except ValueError:
    print("Invalid input. Please enter a number.")

def unmark_complete():
  view_list()
  try:
    task_index = int(input("Enter the index of the task to unmark as complete: ")) - 1
    if 0 <= task_index < len(tasks):
      tasks[task_index]["completed"] = False
      print("Task unmarked as complete!")
    else:
      print("Invalid task index.")
  except ValueError:
      print("Invalid input. Please enter a number.")


while True:
  print("\nChoose an action:")
  print("1. Add task")
  print("2. Delete task")
  print("3. View list")
  print("4. Mark as complete")
  print("5. Unmark as complete")
  print("6. Exit")

  choice = input("Enter your choice (1-6): ")

  if choice == '1':
    add_task()
  elif choice == '2':
    delete_task()
  elif choice == '3':
    view_list()
  elif choice == '4':
    mark_complete()
  elif choice == '5':
    unmark_complete()
  elif choice == '6':
    print("Exiting the to-do list builder.")
    break
  else:
    print("Invalid choice. Please enter a number between 1 and 6.")
```
